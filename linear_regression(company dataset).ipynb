# Linear regression on company dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
advertising = pd.read_csv("company.csv")
advertising
```

```python
print("Shape of our dataset")
print(advertising.shape)
```

```python
print("Info of our dataset")
print(advertising.info())
```

```python
print("Describe our dataset")
print(advertising.describe())
```

```python
sns.pairplot(advertising, x_vars=['TV', 'Radio', 'Newspaper'], 
            y_vars = 'Sales', height = 4, aspect = 1, kind = 'scatter')
plt.show()
```

```python
X = advertising.iloc[:, :-1].values #it will give all the columns
                                    #expect the last columns
Y = advertising.iloc[:, 1].values   #it will give you the last column
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7,
                                        test_size = 0.3, random_state = 100)
print("Training data in:")
print(X_train)
print("Training data out:")
print(Y_train)
print("Testing data in:")
print(X_test)
print("Testing data out:")
print(Y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
Y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
print(df)
```

```python
print("\nRegressor Intercept")
print(regressor.intercept_)
print("\nRegressor Coefficient")
print(regressor.coef_)
print("\nRegressor Score")
print(regressor.score(X, Y))
```
