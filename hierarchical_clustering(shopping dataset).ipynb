### Hierarchical Clustering on shopping dataset using agglomerative method.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customer_data = pd.read_csv("shopping-data.csv")
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 2:5].values
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize = (10, 6))
plt.title("Customer Dendrogram")
dend = shc.dendrogram(shc.linkage(data, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', 
                                 linkage = 'ward')
cluster.fit_predict(data)
print(cluster.labels_)
```

```python
plt.figure(figsize = (9, 6))
plt.scatter(data[:, 1], data[:, 2], c = cluster.labels_, cmap = 'rainbow')
```
