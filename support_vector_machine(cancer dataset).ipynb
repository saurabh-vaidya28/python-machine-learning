<h3>Support Vector Machine on cancer dataset</h3>

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.feature_names)
print('target', cancer.target_names)
```

```python
cancer.data.shape
```

```python
cancer.data[0:5]
```

```python
cancer.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    test_size = 0.3, random_state = 123)
```

```python
cancer.target
```

```python
print("Training data out: \n", y_train)
print("Testing data out: \n", y_test)
```

```python
from sklearn import svm
clf = svm.SVC(kernel = 'linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("The accuracy is", metrics.accuracy_score(y_test, y_pred))
```

```python
print("Precision Score", metrics.precision_score(y_test, y_pred))
```

```python
print("Recall score", metrics.recall_score(y_test, y_pred))
```

```python
import pandas as pd
df = pd.DataFrame({'test':y_test,'predicted':y_pred})
df
```

```python
from pandas_ml import ConfusionMatrix
confusion_matrix=ConfusionMatrix(df['test'],df['predicted'])
print(confusion_matrix)
```

```python
print(confusion_matrix.print_stats())
```

Accuracy of the confusion matrix is 97%

```python
print(confusion_matrix.sensitivity)
```

As we can see sensitivity of confusion matrix is 99%
